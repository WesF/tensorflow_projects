```python
from tensorflow import keras
import tensorflow as tf
from keras.models import Sequential, Model
from keras.layers import Dense, Activation, Input, Flatten, LSTM, BatchNormalization, Dropout
from keras.layers import Embedding, Dropout, Conv3D, Conv1D, MaxPooling1D, LocallyConnected1D
from keras.layers import GlobalAveragePooling1D
from keras.layers.recurrent import LSTM
from keras.utils.np_utils import to_categorical
from keras.callbacks import History
import pandas as pd
import numpy as np
from keras.losses import mse, mae
from keras.activations import relu, selu
from keras import regularizers
import protabank as pb
import talos as ta
import seaborn as sns
import sklearn
```

```python
%run env.py
connection = pb.Connection(username, password)
study = connection.get_study('CdSscVh64')
```

```python
study.encode(features=['code', 'charge', 'hydrophobicity', 'volume'],
              normalize=True, diff_from_wildtype=True)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(study.encoded_sequences,
                                                    study.labels[1:].astype('float'), test_size=0.2)
```

```python
import keras.backend as K

def mse_nan(y_true, y_pred):
    index = ~K.tf.is_nan(y_true)
    y_true = K.tf.boolean_mask(y_true, index)
    y_pred = K.tf.boolean_mask(y_pred, index)
    return K.mean((y_true - y_pred) ** 2)

def multilayer_multi_dense(shape):
    dim = shape[1]*shape[2]
    model = Sequential()
    model.add(Flatten(input_shape=[shape[1], shape[2]]))
    model.add(Dense(100, activation='relu'))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(25, activation='relu'))
    model.add(Dense(5))
    model.compile(loss=mse_nan, optimizer='adam')
    return model
```

```python
from livelossplot import PlotLossesKeras
from keras.callbacks import EarlyStopping

model = multilayer_multi_dense(shape=x_train.shape)
early_stop = EarlyStopping(monitor='val_loss', min_delta=.005, patience=5, verbose=1, mode='auto')
history = model.fit(x_train, y_train, epochs=20, batch_size=128, validation_data=(x_test, y_test),
                               callbacks=[PlotLossesKeras(),],
                                          verbose=0)
predictions = model.predict(x_test)
print("Fitness MSE {}".format(history.history['val_loss'][-1]))
```
